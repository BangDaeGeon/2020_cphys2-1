# 2.Diabetes

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
```

```python
dia = load_diabetes()
print(type(dia))
print(dia.keys())
print(dia.feature_names)
```

```python
x=dia.data
y=dia.target
```

```python
features = dia.feature_names
fig, ax = plt.subplots(3, 4)
print('fig: ', fig)
ax_flat = ax.flatten()
for i in range(len(features)):
    subplot = ax_flat[i]
    subplot.scatter(x[:, i], y)
    subplot.set_title(features[i])
plt.plot();
```

```python
reg=LinearRegression()
lasso_reg=Lasso(alpha=5)
ridge_reg=Ridge(alpha=8)

reg.fit(x,y)
lasso_reg.fit(x,y)
ridge_reg.fit(x,y)
```

```python
for a, b in zip(dia.feature_names, reg.coef_):
    print("%s: %f" %(a,b))
```

```python
for a, b in zip(dia.feature_names, lasso_reg.coef_):
    print("%s: %f" %(a,b))
```

```python
for a, b in zip(dia.feature_names, ridge_reg.coef_):
    print("%s: %f" %(a,b))
```

```python
reg.score(x,y)
```

```python
X=[i[0] for i in x]
plt.title('age',size=20)
plt.plot(X,y,'o')
```

```python
model_linear = LinearRegression().fit(dia.data[:,0].reshape(-1,1),dia.target)
```

```python
plt.scatter(dia.data[:,0],dia.target)
plt.scatter(dia.data[:,0],model_linear.predict(dia.data[:,0].reshape(-1,1)))
plt.title('age',size=20)
```

```python
model_linear.score(dia.data[:,0].reshape(-1,1),dia.target)
```

```python
poly_2 = PolynomialFeatures(degree=2)
poly_2_X = poly_2.fit_transform(dia.data[:,0].reshape(-1,1))
 
model_2_poly = LinearRegression().fit(poly_2_X,dia.target)
plt.scatter(dia.data[:,0],dia.target,alpha=0.25)
 
linear_fit = plt.plot(dia.data[:,0], model_linear.predict(dia.data[:,0].reshape(-1,1)), color = 'red',linestyle='-')
poly_2_fit = plt.scatter(dia.data[:,0], model_2_poly.predict(poly_2_X), color = 'green', linestyle='--')
plt.title('age',size=20)
```

```python
model_2_poly.score(poly_2_X,dia.target)
```

```python
poly_3 = PolynomialFeatures(degree=3)
poly_3_X = poly_3.fit_transform(dia.data[:,0].reshape(-1,1))
 
model_3_poly = LinearRegression().fit(poly_3_X,dia.target)
plt.scatter(dia.data[:,0],dia.target,alpha=0.25)
 
linear_fit = plt.plot(dia.data[:,0], model_linear.predict(dia.data[:,0].reshape(-1,1)), color = 'red')
poly_2_fit = plt.scatter(dia.data[:,0], model_2_poly.predict(poly_2_X), color = 'green')
poly_3_fit = plt.scatter(dia.data[:,0], model_3_poly.predict(poly_3_X), color = 'cyan')
plt.title('age',size=20)
```

```python
model_3_poly.score(poly_3_X,dia.target)
```
