## Logistic Regression

### 1.꽃받침 길이에 따른 virginica 분류

```python
X0=[[i] for i in iris.data[:,0]] # 꽃받침 길이
y0=[1 if i==2 else 0 for i in iris.target] # virginica이면 1, 아니면 0
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression(solver='lbfgs')
log_reg.fit(X0,y0)
```

```python
x0=[[i] for i in np.linspace(4,8,100)]
plt.plot(X0,y0,'o')
plt.plot(x0,log_reg.predict(x0))
```

```python
log_reg.score(X0,y0)
```

### 2.꽃받침 넓이에 따른 virginica 분류

```python
X1=[[i] for i in iris.data[:,1]] # 꽃받침 넓이
y1=[1 if i==2 else 0 for i in iris.target] # virginica이면 1, 아니면 0
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression(solver='lbfgs')
log_reg.fit(X1,y1)
```

```python
x1=[[i] for i in np.linspace(1,8,100)]
plt.plot(X1,y1,'o')
plt.plot(x1,log_reg.predict(x1))
```

```python
log_reg.score(X1,y1)
```

### 3.꽃잎 길이에 따른 virginica 분류

```python
X2=[[i] for i in iris.data[:,2]] # 꽃잎 길이
y2=[1 if i==2 else 0 for i in iris.target] # virginica이면 1, 아니면 0
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression(solver='lbfgs')
log_reg.fit(X2,y2)
```

```python
x2=[[i] for i in np.linspace(0,8,100)]
plt.plot(X2,y2,'o')
plt.plot(x2,log_reg.predict(x2))
```

```python
log_reg.score(X2,y2)
```

### 4.꽃잎 넓이에 따른 virginica 분류

```python
X3=[[i] for i in iris.data[:,3]] # 꽃잎 넓이
y3=[1 if i==2 else 0 for i in iris.target] # virginica이면 1, 아니면 0
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression(solver='lbfgs')
log_reg.fit(X3,y3)
```

## Support Vector Machine

```python
from sklearn.svm import SVC 

svm=SVC(kernel='rbf',gamma='auto')
svm.fit(X2,y2)
plt.scatter(X2,y2,c=svm.predict(X2))
svm.score(X2,y2)
```

```python
X=iris["data"] 
y=iris["target"] # setosa=0, versicolor=1, virginica=2
```

```python
svm=SVC(kernel='rbf',gamma='auto')
svm.fit(X,y)
svm.score(X,y)
plt.scatter(X[:,0],X[:,1],c=svm.predict(X))
```

# 보고서

 이번 코딩의 목적은 파이선의 Iris 데이터를 이용하여 Logistic 회귀, Support Vector Machine 방법을 이용하여 분석하는 것이다.
 
 Iris (분꽃)에는 'setosa', 'versicolor', 'virginica'의 3가지 종류의 분꽃(target)이 있는데 각각의 꽃받침, 꽃잎의 길이, 넓이(data)가 다르기 때문에 파이썬 내의 데이터를 이용하여 Logistic Regression으로 data에 따른 target을 세세히 분류 할 수 있으며, Support Vector Machine중에서도 kernel trick을 이용하여 저차원 공간을 고차원 공간으로 매핑하여 간결하게 표시할 수 있다.
 
 Logistic Regression를 사용하여 data에 따른 virginica를 구별한 결과는 꽃받침 길이의 경우 0.8score 즉 80.0%로 들어 맞았고, 꽃받침 넓이에 따른 분석은 66.6%로 virginica였으며, 꽃잎의 길이와 넓이에 따른 결과는 각각 95.3%, 96.0%로 virginica를 분류 할 수 있었다.
 
 Support Vector Machine을 이용하여 구한 꽃잎의 길이에 따른 virginica의 분류도 95.3%로 Logistic 회귀에 따른 결과와 같은걸 보이며 꽃잎의 길이에 따른 결과 뿐만이 아니라 꽃받침의 길이, 넓이, 꽃잎의 넓이 또한 Logistic 회귀의 결과와 같을 것을 추측 할 수 있으며 Logistic 회귀보다 이미지를 더 간결하게 볼 수 있었다. 또, kernel방식을 이용하여 3가지 종류의 분꽃을 구별하여 스캐터링 하였을 때 산점도 그래프를 볼 수 있으며 이는 98.6%로 분류 되었음을 볼 수 있었다.
