# 1.Boston House Price

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
```

```python
boston = load_boston()
print(type(boston))
print(boston.keys())
print(boston.feature_names)
```

```python
x=boston.data
y=boston.target
```

### 'CRIM' 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO',
### 'B', 'LSTAT' 의 산점도 그래프

```python
features = boston.feature_names
fig, ax = plt.subplots(3, 5)
print('fig: ', fig)
ax_flat = ax.flatten()
for i in range(len(features)):
    subplot = ax_flat[i]
    subplot.scatter(x[:, i], y)
    subplot.set_title(features[i])
plt.plot();
```

```python
reg=LinearRegression()
lasso_reg=Lasso(alpha=5)
ridge_reg=Ridge(alpha=8)

reg.fit(x,y)
lasso_reg.fit(x,y)
ridge_reg.fit(x,y)
```

```python
for a, b in zip(boston.feature_names, reg.coef_):
    print("%s: %f" %(a,b))
```

```python
for a, b in zip(boston.feature_names, lasso_reg.coef_):
    print("%s: %f" %(a,b))
```

```python
for a, b in zip(boston.feature_names, ridge_reg.coef_):
    print("%s: %f" %(a,b))
```

```python
reg.score(x,y)
```

### 'LSTAT' 의 산점도 그래프

```python
X=[i[12] for i in x]
plt.title('Price ~ LSTAT',size=20)
plt.xlabel('LSTAT')
plt.ylabel('PRICE')
plt.plot(X,y,'o')
```

### 'LSTAT' 의 회귀 분석

```python
model_linear = LinearRegression().fit(boston.data[:,12].reshape(-1,1),boston.target)
```

```python
plt.scatter(boston.data[:,12],boston.target)
plt.scatter(boston.data[:,12],model_linear.predict(boston.data[:,12].reshape(-1,1)))
plt.title('Price ~ LSTAT',size=20)
plt.xlabel('LSTAT')
plt.ylabel('PRICE')
```

```python
model_linear.score(boston.data[:,12].reshape(-1,1),boston.target)
```

```python
poly_2 = PolynomialFeatures(degree=2)
poly_2_X = poly_2.fit_transform(boston.data[:,12].reshape(-1,1))
 
model_2_poly = LinearRegression().fit(poly_2_X,boston.target)
plt.scatter(boston.data[:,12],boston.target,alpha=0.25)
 
linear_fit = plt.plot(boston.data[:,12], model_linear.predict(boston.data[:,12].reshape(-1,1)), color = 'red',linestyle='-')
poly_2_fit = plt.scatter(boston.data[:,12], model_2_poly.predict(poly_2_X), color = 'green', linestyle='--')
plt.title('Price ~ LSTAT',size=20)
plt.xlabel('LSTAT')
plt.ylabel('PRICE')
```

```python
model_2_poly.score(poly_2_X,boston.target)
```

```python
poly_3 = PolynomialFeatures(degree=3)
poly_3_X = poly_3.fit_transform(boston.data[:,12].reshape(-1,1))
 
model_3_poly = LinearRegression().fit(poly_3_X,boston.target)
plt.scatter(boston.data[:,12],boston.target,alpha=0.25)
 
linear_fit = plt.plot(boston.data[:,12], model_linear.predict(boston.data[:,12].reshape(-1,1)), color = 'red')
poly_2_fit = plt.scatter(boston.data[:,12], model_2_poly.predict(poly_2_X), color = 'green')
poly_3_fit = plt.scatter(boston.data[:,12], model_3_poly.predict(poly_3_X), color = 'cyan')
plt.title('Price ~ LSTAT',size=20)
plt.xlabel('LSTAT')
plt.ylabel('PRICE')
```

```python
model_3_poly.score(poly_3_X,boston.target)
```

```python
poly_10 = PolynomialFeatures(degree=10)
poly_10_X = poly_10.fit_transform(boston.data[:,12].reshape(-1,1))
 
model_10_poly = LinearRegression().fit(poly_10_X,boston.target)
plt.scatter(boston.data[:,12],boston.target,alpha=0.25)
 
linear_fit = plt.plot(boston.data[:,12], model_linear.predict(boston.data[:,12].reshape(-1,1)), color = 'red')
poly_2_fit = plt.scatter(boston.data[:,12], model_2_poly.predict(poly_2_X), color = 'green')
poly_10_fit = plt.scatter(boston.data[:,12], model_10_poly.predict(poly_10_X), color = 'cyan')
plt.title('Price ~ LSTAT',size=20)
plt.xlabel('LSTAT')
plt.ylabel('PRICE')
```
